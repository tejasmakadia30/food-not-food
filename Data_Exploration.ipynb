## Food List from NLTK

```python
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet as wn
food = wn.synset("food.n.02")
food_classes = list(set([w for s in food.closure(lambda s:s.hyponyms()) for w in s.lemma_names()]))
```

```python
food_classes
```

```python
food_lst = [food_item.lower().split('_') for food_item in food_classes]
food_lst
```

```python
flat_food_lst = [food for food_sub_lst in food_lst for food in food_sub_lst]
flat_food_lst
```

```python
imagenet_food_classes = {}
for k, v in imagenet_classes.items(): #look for imagenet classes
  imagenet_class_set = set([space_word.strip(" ") for space_word in v.lower().split(",")]) # get value from imagenet classes
  if imagenet_class_set.intersection(flat_food_lst): # check if it appears in flat_food_lst
    imagenet_food_classes[k] = v
    # print(k,v)
imagenet_food_classes
```

```python
nonfood_classes_manual_sort = [457,494,567,626,723,738,760,923,972,976]
```

```python
imagenet_manual_filtered_food_classes = {}
for k,v in imagenet_food_classes.items():
  if k not in nonfood_classes_manual_sort:
    imagenet_manual_filtered_food_classes[k] = v
imagenet_manual_filtered_food_classes
```

```python
food_classes_keys = list(imagenet_manual_filtered_food_classes.keys())
food_classes_keys[:10]
```

```python
# get list of food and non food classes from imagenet
imagenet_non_food_classes = {}
for k,v in imagenet_classes.items():
  if k not in food_classes_keys:
    imagenet_non_food_classes[k] = v

len(imagenet_non_food_classes)
```

```python
# Move food images from ImageNet downloaded folders to /food_images
from shutil import copy2
start_dir = "/content/drive/MyDrive/ML/Food-Not-Food/images/imagenet_images/imagenet_images"
dest_dir = "/content/drive/MyDrive/ML/Food-Not-Food/images/food-images"
for image_folder in os.listdir(start_dir):
  target_image_folder = None
  if image_folder.lower() in food_list:
    new_dest_dir = os.path.join(dest_dir, image_folder)
    os.makedirs(new_dest_dir, exist_ok=True)

    target_image_folder = os.path.join(start_dir, image_folder)
    
  else: pass

  if target_image_folder:
      images_to_copy = os.listdir(target_image_folder)
      for image_to_copy in images_to_copy:
        image_filename = image_to_copy.split()[-1]
        start_path = os.path.join(target_image_folder, image_filename)
        dest_path = os.path.join(new_dest_dir, image_filename)
        print(f"Copying: {start_path} to {dest_path}...")
        copy2(start_path, dest_path)
```

```python
# Move food images from ImageNet downloaded folders to /food_images
from shutil import copy2
count = 0
start_dir = "/content/drive/MyDrive/ML/Food-Not-Food/images/imagenet_images/imagenet_images"
dest_dir = "/content/drive/MyDrive/ML/Food-Not-Food/images/non-food-images"
for image_folder in os.listdir(start_dir):
  target_image_folder = None
  if image_folder.lower() in non_food_list_filter:
    new_dest_dir = os.path.join(dest_dir, image_folder)
    os.makedirs(new_dest_dir, exist_ok=True)

    target_image_folder = os.path.join(start_dir, image_folder)
    
  else: pass

  if target_image_folder:
      images_to_copy = os.listdir(target_image_folder)
      for image_to_copy in images_to_copy:
        image_filename = image_to_copy.split()[-1]
        start_path = os.path.join(target_image_folder, image_filename)
        dest_path = os.path.join(new_dest_dir, image_filename)
        print(f"Copying: {start_path} to {dest_path}...")
        copy2(start_path, dest_path)
        count += 1
print(count)
```
